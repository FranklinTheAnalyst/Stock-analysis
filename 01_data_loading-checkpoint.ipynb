```python
# Select one stock (you can loop over multiple later)
ticker = df['ticker'].unique()[0]  # change this to visualize another
stock_data = df1[df1['ticker'] == ticker].sort_values('date')

# Calculate daily return and rolling mean to smooth noise
stock_data['return'] = stock_data['close'].pct_change()
stock_data['rolling_mean'] = stock_data['close'].rolling(window=10).mean()

# Define trend phases based on rolling slope
# If slope > threshold = uptrend, < -threshold = downtrend, otherwise sideways
window = 5
stock_data['slope'] = stock_data['rolling_mean'].diff(window)
threshold = stock_data['slope'].std() * 0.5  # adaptive threshold
stock_data['trend'] = np.where(stock_data['slope'] > threshold, 'Uptrend',
                        np.where(stock_data['slope'] < -threshold, 'Downtrend', 'Sideways'))

# Plot the closing prices with colored trend regions
plt.figure(figsize=(12,6))
for trend, color in zip(['Uptrend', 'Downtrend', 'Sideways'], ['blue', 'red', 'yellow']):
    trend_data = stock_data[stock_data['trend'] == trend]
    plt.plot(trend_data['date'], trend_data['close'], color=color, label=trend)

plt.title(f'Visual Trend Identification for {ticker}')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
macd = ta.trend.MACD(df1['close'])
df1['MACD'] = macd.macd()
df1['Signal'] = macd.macd_signal()
df1['MACD_Hist'] = macd.macd_diff()
```

### Bollinger Bands

```python
bb = ta.volatility.BollingerBands(df1['close'], window=20, window_dev=2)
df1['BB_High'] = bb.bollinger_hband()
df1['BB_Low'] = bb.bollinger_lband()
df1['BB_Mid'] = bb.bollinger_mavg()
```

### Bollinger Bands: Identify Price Volatility Patterns

```python
plt.figure(figsize=(12,6))
plt.plot(df1['close'], label='close Price', color='black')
plt.plot(df1['BB_Mid'], label='Middle Band', color='blue')
plt.plot(df1['BB_High'], label='Upper Band', color='red', linestyle='--')
plt.plot(df1['BB_Low'], label='Lower Band', color='green', linestyle='--')
plt.fill_between(df1.index, df1['BB_Low'], df1['BB_High'], color='gray', alpha=0.1)
plt.title('Bollinger Bands')
plt.legend()
plt.show()
```

### Insights Bollinger Bands

```python
The middle band (blue) represents the moving average, capturing the overall trend.

The upper (red) and lower (green) bands expand and contract depending on price volatility.

When the bands widen, volatility is increasing; when they narrow, volatility is decreasing.

Price peaks near the upper band often indicate overbought conditions, while touches near the lower band suggest oversold conditions.
```

```python
# Technical indicators
df1['RSI'] = ta.momentum.RSIIndicator(df1['close'], window=14).rsi()
df1['MACD'] = ta.trend.MACD(df1['close']).macd()
df1['Signal'] = ta.trend.MACD(df1['close']).macd_signal()
df1['BB_Width'] = (ta.volatility.BollingerBands(df1['close']).bollinger_hband() -
                    ta.volatility.BollingerBands(df1['close']).bollinger_lband())
df1['SMA_20'] = df1['close'].rolling(20).mean()
df1['SMA_50'] = df1['close'].rolling(50).mean()
df1['Returns'] = df1['close'].pct_change()


```

```python
# Drop missing values (from indicator calculations)
df1.dropna(inplace=True)
```

### Correlation Matrix

```python
corr_matrix = df1[['close', 'RSI', 'MACD', 'Signal', 'BB_Width', 'SMA_20', 'SMA_50', 'Returns']].corr()
corr_matrix
# numerical correlation coefficients between features.
```

## Visualize as a Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Technical Indicators')
plt.show()
```

### Insights on heatmap

```python
The Close price, SMA_20, and SMA_50 are strongly correlated (values around 0.95â€“0.98). Thatâ€™s expected since moving averages are derived directly from price.


MACD and Signal lines are also highly correlated (0.96) â€” meaning they often move together, which can create redundancy if both are used in modeling.


RSI shows moderate correlation with MACD and Signal (around 0.6), suggesting it captures a slightly different momentum pattern.   
```
